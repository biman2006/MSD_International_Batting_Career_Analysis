## TOTAL RUNS SCORED BY MS  IN HIS INTERNATIONAL CAREER

```python
query=""" 

SELECT SUM(Runs) AS Total_Runs FROM ms_dhoni;

"""
Total_Runs=pd.read_sql(query,conn)
display(Total_Runs)
```

## TOP 5 HIGHEST INDIVIDUAL SCORE OF MSD IN HIS CAREER

```python
query=""" 
SELECT Runs AS Highest_Individual_Score FROM ms_dhoni
ORDER BY Runs DESC
LIMIT 5;


"""

Highest_Individual_Score=pd.read_sql(query,conn)
display(Highest_Individual_Score)
```

## TOTAL CENTURY AND HALF CENTURY OF MSD IN HIS INTERNATIONAL CAREER

```python
query="""
SELECT
  SUM(CASE WHEN Runs>=100 THEN 1 ELSE 0 END) AS centuries,
  SUM(CASE WHEn Runs BETWEEN 50 AND 99 THEN 1 ELSE 0 END) AS Half_Centuries
FROM ms_dhoni;

"""

Total_Century_AND_Half_Centuries=pd.read_sql(query,conn)
display(Total_Century_AND_Half_Centuries)
```

## TOTAL NUMBER OF NOT OUT INNS 

```python
query="""  
SELECT COUNT(*) AS Total_Not_Out FROM ms_dhoni
WHERE Dismissal= "Not Out";

"""

Total_Not_Out=pd.read_sql(query,conn)
display(Total_Not_Out)
```

## TOTAL RUNS SCORED PER YEAR

```python
query=""" 
SELECT YEAR(`Date`) AS year,
SUM(Runs) AS total_Runs
FROM ms_dhoni
GROUP BY year
ORDER BY year;

"""

Runs_Per_Year=pd.read_sql(query,conn)
display(Runs_Per_Year)
```
